We define our functions as:

```python
def f_4a(x):
    return 2*x+1

def f_4b(x):
    return x*x*x
```

Our load assembler is:

```python

#COPIED FROM DR KAMAN'S SAMPLE PROGRAMS

def LoadAssembler(x,f):
    n = len(x)                                      #we create a variable to represent length of n
    b = np.zeros((n,1))                             #initialize vector b to be the same length as x
    for i in range(0,n-1):                          #loop over i that goes to every entry in b and calculate
        h = x[i+1]-x[i]                             #h is the distance from this x[i] to the next
        b[i] = b[i] + f(x[i]) * (h/2)               #each entry in b is f of entry at i + h/2
        b[i+1]= b[i+1] + f(x[i+1]) * (h/2)          #we calculate the next b 
    return b                                        #return b
```

Our mass assembler is:

```python
#COPIED FROM DR KAMAN'S SAMPLE PROGRAMS

def MassAssembler(x):
    M = np.zeros((len(x),len(x)))                   #initialize m as an nxn matrix
    n = len(x)                                      #n is as long as x
    for i in range(0,n-1):                          #we iterate to fill out the matrix
        h = x[i+1]-x[i]                             #h is the distance from this x[i] to the next
        M[i,i] = M[i,i] + (h/3)                     #the diagonals are all h/3
        M[i,i+1] = M[i,i+1] +(h/6)                  #the superdiagonals are h/6
        M[i+1,i] = M[i+1,i] +(h/6)                  #the subdiagonals are h/6
        M[i+1,i+1] = M[i+1,i+1] + (h/3)             #we calculate the next diagonal value
    return M                                        #return M
```

Our main function for number 4a and 4b is:

```python
x = np.linspace(-1,1,10)                                        #create 1D vector

M= MassAssembler(x)                                             #We calculate the mass matrix
b = LoadAssembler(x,f_4a)                                       #we calculate load vector

Phf = np.linalg.inv(M).dot(b)                                   #solve the linear system

plt.plot(x,f_4a(x))                                             #we plot function for 4a
plt.plot(x,Phf,'--',marker = 'o')                               #use -- and markers o for the projection
plt.legend(["f(x)","Projection of f(x)"])                       #create legend
plt.xlabel("x")                                                 #label axes
plt.ylabel("y")
plt.title("f(x) vs L2 projection of f(x)")                      #create title
plt.show
```
